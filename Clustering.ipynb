**Clustering based on PCA result**

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

import numpy as np
```

```python
# Load PCA and AE data
df_pca = pd.read_csv("./PCA/car_price_pca.csv")
df_ae = pd.read_csv("./AE/car_price_ae.csv") 
df_sae = pd.read_csv("./SAE/car_price_sae.csv")

# Function to perform clustering and visualization
def cluster_and_visualize(data, name, x_col, y_col):
    ks = range(2, 15)
    sil_scores = []
    db_scores = []  # Davies-Bouldin scores
    inertias = []   # Elbow method scores
    score_dict = {}
    
    # Calculate multiple metrics for each k
    for k in ks:
        km = KMeans(n_clusters=k, random_state=42)
        labels = km.fit_predict(data)
        
        # Silhouette score (higher is better)
        sil = silhouette_score(data, labels)
        sil_scores.append(sil)
        
        # Davies-Bouldin score (lower is better)
        db = davies_bouldin_score(data, labels)
        db_scores.append(db)
        
        # Inertia for elbow method (lower is better)
        inertias.append(km.inertia_)
        
        print(f"{name} - k = {k}:")
        print(f"  Silhouette Score = {sil:.4f}")
        print(f"  Davies-Bouldin Score = {db:.4f}")
        print(f"  Inertia = {km.inertia_:.4f}")
        score_dict[k] = sil, db, km.inertia_


    # Normalize scores to 0-1 range for comparison
    norm_sil = (sil_scores - np.min(sil_scores)) / (np.max(sil_scores) - np.min(sil_scores))
    norm_db = 1 - (db_scores - np.min(db_scores)) / (np.max(db_scores) - np.min(db_scores))  # Invert DB score
    norm_inertia = 1 - (inertias - np.min(inertias)) / (np.max(inertias) - np.min(inertias))  # Invert inertia
    
    # Combined score (weighted average)
    combined_scores = 0.4 * norm_sil + 0.3 * norm_db + 0.3 * norm_inertia

    # Visualize all metrics
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    ax1.plot(list(ks), sil_scores, marker='o')
    ax1.set_xlabel("Number of clusters (k)")
    ax1.set_ylabel("Silhouette Score")
    ax1.set_title("Silhouette Analysis")
    ax1.grid(True)
    
    ax2.plot(list(ks), db_scores, marker='o')
    ax2.set_xlabel("Number of clusters (k)")
    ax2.set_ylabel("Davies-Bouldin Score")
    ax2.set_title("Davies-Bouldin Analysis")
    ax2.grid(True)
    
    ax3.plot(list(ks), inertias, marker='o')
    ax3.set_xlabel("Number of clusters (k)")
    ax3.set_ylabel("Inertia")
    ax3.set_title("Elbow Method")
    ax3.grid(True)
    
    ax4.plot(list(ks), combined_scores, marker='o')
    ax4.set_xlabel("Number of clusters (k)")
    ax4.set_ylabel("Combined Score")
    ax4.set_title("Combined Metric")
    ax4.grid(True)
    
    plt.tight_layout()
    plt.savefig(f"{name}/clustering_metrics_{name.lower()}.png", dpi=300)
    plt.show()

    optimal_k = ks[np.argmax(combined_scores)]
    print(f"\n{name} - Optimal k from combined metrics: {optimal_k}")
    print(score_dict[optimal_k])
    
    # Assess cluster stability
    n_runs = 5
    stability_scores = []
    for _ in range(n_runs):
        kmeans = KMeans(n_clusters=optimal_k, random_state=np.random.randint(0, 1000))
        labels = kmeans.fit_predict(data)
        stability_scores.append(silhouette_score(data, labels))
    
    print(f"Cluster stability (std of silhouette scores across {n_runs} runs): {np.std(stability_scores):.4f}")

    # Final clustering with optimal k
    kmeans = KMeans(n_clusters=optimal_k, random_state=42)
    data['cluster'] = kmeans.fit_predict(data)

    # Original scatter plot
    plt.figure(figsize=(10, 10))
    cmap = get_cmap("tab10", optimal_k)
    scatter = plt.scatter(
        data[x_col], data[y_col],
        c=data['cluster'], 
        cmap=cmap,
        s=50, alpha=0.8, edgecolors='k'
    )
    
    handles, labels = scatter.legend_elements(prop="colors")
    plt.legend(handles, [f"Cluster {i}" for i in range(optimal_k)],
               title="Clusters", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f"K-Means Clustering on {name}-reduced Data")
    plt.tight_layout()
    plt.savefig(f"{name}/kmeans_{name.lower()}_scatter.png", dpi=300)
    plt.show()

    # t-SNE visualization
    tsne_data = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000).fit_transform(data.drop(columns=['cluster']))
    
    plt.figure(figsize=(10, 10))
    for i in range(optimal_k):
        mask = data['cluster'] == i
        plt.scatter(tsne_data[mask, 0], tsne_data[mask, 1], label=f'Cluster {i}')
    plt.title(f't-SNE visualization of KMeans clustering - {name}')
    plt.legend()
    plt.savefig(f"{name}/tsne_kmeans_{name.lower()}.png", dpi=300)
    plt.show()
    
    return data['cluster']

# Perform clustering on PCA data
print("Clustering PCA data:")
pca_clusters = cluster_and_visualize(df_pca, "PCA", "PC1", "PC2")

# Perform clustering on AE data  
print("\nClustering AE data:")
ae_clusters = cluster_and_visualize(df_ae, "AE", "AE_1", "AE_2")

# Perform clustering on SAE data  
print("\nClustering SAE data:")
sae_clusters = cluster_and_visualize(df_sae, "SAE", "SAE_1", "SAE_2")
```

```python
def find_competitors(df_orig, df_clusters, name):
    df_orig['cluster'] = df_clusters['cluster']
    
    vw_cluster = df_orig.loc[df_orig['CarBrand_volkswagen'] == 1, 'cluster'].mode()[0]
    print(f"Cluster number for Volkswagen ({name}): {vw_cluster}")
    
    competitors = df_orig[df_orig['cluster'] == vw_cluster]
    
    competitors.to_csv(f"{name}/vw_competitors_{name.lower()}.csv", index=False)
    print(f"Found {len(competitors)} competitor models for {name}, saved to vw_competitors_{name.lower()}.csv")
    return competitors

df_orig = pd.read_csv("./Data/car_price_preprocessed.csv")
pca_competitors = find_competitors(df_orig.copy(), df_pca, "PCA")

ae_competitors = find_competitors(df_orig.copy(), df_ae, "AE")

sae_competitors = find_competitors(df_orig.copy(), df_sae, "SAE")
```
